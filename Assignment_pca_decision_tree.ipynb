# Splitting the dataset into training and testing dataset

```python

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(data, label_of_data, test_size=0.3, random_state=101, shuffle = True)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import accuracy_score,classification_report, precision_recall_curve, confusion_matrix
decision_tree = DecisionTreeClassifier(max_depth = 3 , min_samples_split = 5)
decision_tree.fit(X_train, Y_train)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = round(decision_tree.score(X_test, Y_test) * 100, 2)
print('Accuracy is: ',acc_decision_tree)
print(classification_report(Y_test, Y_pred))
```

# Changing the max_depth parameter and evaluating Training and Testing Accuracy

```python
tree_depth = 30
train_accuracy = np.empty((tree_depth,1))
test_accuracy = np.empty((tree_depth,1))

# Loop over different values of k
for i in range(0,tree_depth):
    # Setup a Decision Tree Classifier
    decision_tree = DecisionTreeClassifier(max_depth = i+1)

    # Fit the classifier to the training data
    decision_tree.fit(X_train, Y_train)

    #Compute accuracy on the training set
    train_accuracy[i] = decision_tree.score(X_train, Y_train)

    #Compute accuracy on the testing set
    test_accuracy[i] = decision_tree.score(X_test, Y_test)

# Generate plot
plt.title('decision_tree: Varying depth of tree')
plt.plot(test_accuracy, label = 'Testing Accuracy')
plt.plot(train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Depth of tree')
plt.ylabel('Accuracy')
plt.show()
```

Looking at the graph above we can see that, approximately, there is no any noticeable overfitting when max_depth is upto 30.

# Calculating eigen values and eigen vectors of Covariance Matrix

```python
cov_mat = np.cov(data_std.T)
eigen_vals, eigen_vecs = np.linalg.eig(cov_mat)
print('\nEigenvalues \n%s' % eigen_vals)
```

```python
eigen_vals.shape, eigen_vecs.shape
```

```python
tot = sum(eigen_vals)
var_exp = [(i / tot) for i in sorted(eigen_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
```

```python
plt.bar(range(1, 179), var_exp, alpha=0.5, align='center',
        label='individual explained variance')
plt.step(range(1, 179), cum_var_exp, where='mid',
         label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
# plt.savefig('images/05_02.png', dpi=300)
plt.show()
```

From the above diagram we can conclude that, approximately, the first 25 sorted eigen values and their corresponding eigen vectors are more significant and contributing.So, they can be considered for transformation in new axis.

```python
# Make a list of (eigenvalue, eigenvector) tuples
eigen_pairs = [(np.abs(eigen_vals[i]), eigen_vecs[:, i])
               for i in range(len(eigen_vals))]

# Sort the (eigenvalue, eigenvector) tuples from high to low
eigen_pairs.sort(key=lambda k: k[0], reverse=True)

w = np.hstack((eigen_pairs[0][1][:, np.newaxis],
               eigen_pairs[1][1][:, np.newaxis],
               eigen_pairs[1][1][:, np.newaxis],
               eigen_pairs[2][1][:, np.newaxis],
               eigen_pairs[3][1][:, np.newaxis],
               eigen_pairs[4][1][:, np.newaxis],
               eigen_pairs[5][1][:, np.newaxis],
               eigen_pairs[6][1][:, np.newaxis],
               eigen_pairs[7][1][:, np.newaxis],
               eigen_pairs[8][1][:, np.newaxis],
               eigen_pairs[9][1][:, np.newaxis],
               eigen_pairs[10][1][:, np.newaxis],
               eigen_pairs[11][1][:, np.newaxis],
               eigen_pairs[12][1][:, np.newaxis],
               eigen_pairs[13][1][:, np.newaxis],
               eigen_pairs[14][1][:, np.newaxis],
               eigen_pairs[15][1][:, np.newaxis],
               eigen_pairs[16][1][:, np.newaxis],
               eigen_pairs[17][1][:, np.newaxis],
               eigen_pairs[18][1][:, np.newaxis],
               eigen_pairs[19][1][:, np.newaxis],
               eigen_pairs[20][1][:, np.newaxis],
               eigen_pairs[21][1][:, np.newaxis],
               eigen_pairs[22][1][:, np.newaxis],
               eigen_pairs[23][1][:, np.newaxis],
               eigen_pairs[24][1][:, np.newaxis]))
print('Matrix W:\n', w)
```

```python
eigen_pairs[1]
```

```python
data_std[0].dot(w)
```

# Applying PCA

```python
X_pca = data_std.dot(w)
colors = ['r', 'b', 'g']
markers = ['s', 'x', 'o']

for l, c, m in zip(np.unique(label_of_data), colors, markers):
    plt.scatter(X_pca[label_of_data== l, 0], 
                X_pca[label_of_data== l, 1], 
                c=c, label=l, marker=m)

plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.legend(loc='lower left')
plt.tight_layout()
# plt.savefig('images/05_03.png', dpi=300)
plt.show()
```

Finally, after the application of PCA the features are very much independent and orthogonal to each other. They seem to be more separable than before.

# Using the direct PCA feature from sklearn

```python
from sklearn.decomposition import PCA

pca = PCA()
X_pca = pca.fit_transform(data_std)
pca.explained_variance_ratio_
```

```python
plt.bar(range(1, 179), pca.explained_variance_ratio_, alpha=0.5, align='center')
plt.step(range(1, 179), np.cumsum(pca.explained_variance_ratio_), where='mid')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.show()
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_pca, label_of_data, test_size=0.3, random_state=101, shuffle = True)
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

# PCA applied data are fed into decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.metrics import accuracy_score,classification_report, precision_recall_curve, confusion_matrix
decision_tree = DecisionTreeClassifier(max_depth = 3 , min_samples_split = 5)
decision_tree.fit(X_train, Y_train)
Y_pred = decision_tree.predict(X_test)
acc_decision_tree = round(decision_tree.score(X_test, Y_test) * 100, 2)
print('Accuracy is: ',acc_decision_tree)
print(classification_report(Y_test, Y_pred))
```

We are able to increase the accuracy from 90.38 to 92.87 after the application of PCA. So the PCA seems to be useful for this dataset.
